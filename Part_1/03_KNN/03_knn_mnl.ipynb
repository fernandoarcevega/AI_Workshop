```python
# Split dataset into training and test sets (80%-20%)
tr, tt = HoldOut(Y, 0.2)

# Training set
Xtr = X[tr, :]
Ytr = Y[tr]

# Test set
Xtt = X[tt, :]
Ytt = Y[tt]
```

```python
# Data normalization
Xtr, stats = zscorenorm(Xtr)
Xtt = zscorenorm(Xtt, stats)
```

The **Euclidean distance** is a commonly used metric to determine the proximity between data points:

$$ d_2(\mathbf{a}, \mathbf{b}) = \left(\sum_{i=1}^{d} |a_i - b_i|^2\right)^{1/2}. $$

The posterior probability is calculated directly as

$$ p(\omega_i|\mathbf{x}) = \frac{k_i}{k}, $$

where $k_i$ is the number of class $\omega_i$ training samples nearest to $\mathbf{x}$.

```python
# Knn algorithm
def kNNpredict(Xtr, Ytr, Xtt, k):
  # Euclidean distance
    dist = lambda a, b: np.sqrt(np.sum(np.abs(a - b) ** 2))

    # Indicator function
    I = lambda y, j: (y == j)

    c = np.unique(Ytr)              # Class labels
    nc = len(c)                     # Number of classes
    Ntr = Xtr.shape[0]              # Number of training samples
    Ntt = Xtt.shape[0]              # Number of test samples
    Tmp = np.zeros(Ntt, dtype=int)  # Temporal array
    Prb = np.zeros((Ntt, nc))       # Posterior probabilities

    for a in range(Ntt):            # Classify each test sample
        D = np.zeros(Ntr)

        for b in range(Ntr):
            D[b] = dist(Xtt[a, :], Xtr[b, :]) # Distances to all training samples

        idx = np.argsort(D)   # Sort distances in ascending order
        V = np.zeros(nc)      # Votes for each class

        for b in range(k):    # Check the class of the k-nearest points
            pos = I(c, Ytr[idx[b]])
            V[pos] += 1       # Vote

        Tmp[a] = np.argmax(V) # Classify the test sample into the majority class
        Prb[a, :] = V / k     # Posterior probability

    Ypp = c[Tmp]              # Predicted class labels
    return Ypp, Prb
```

**k-NN algorithm**:
$$
\begin{align*}
& \textbf{input:} \quad X = \{\mathbf{x}_i, y_i\}_{i=1}^n, \mathbf{x}_t, k \\
\\
1: \quad & \textbf{for each } \mathbf{x}_i \in X \textbf{ do} \\
2: \quad &  \text{Compute distance } d_2(\mathbf{x}_i, \mathbf{x}_t) \\
3: \quad & \text{Select } X_k \subseteq X, \text{ the set of } k\text{-NNs to } \mathbf{x}_t \\
\\
& \textbf{return:} \quad y_t = \operatorname*{arg\,max}_{j=1, \dots, c} \sum_{\mathbf{x}_i \in X_k} \mathbf{1}_j(y_i)
\end{align*}
$$

```python
# Classify with kNN
k = 3
Ypp, Prb = kNNpredict(Xtr, Ytr, Xtt, k)
```

```python
# Print classification error
err = np.mean(Ypp != Ytt)
print(f'Error: {100 * err:.3f}%')
```

```python
# Confusion matrix
cm = confusion_matrix(Ytt, Ypp)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
